### Data Exploration
I'll be doing some basic data exploration by asking some questions and getting my answers from the dataset.

Some of the questions are: 

* The top 5 best selling platforms
* The top 50 best selling games of all time
* Total sales  trend between 1980 to 2020.

### The top 5 best selling platforms

```python
selling_platform = df[['Platform', 'Global_Sales']].groupby(
    ['Platform']).sum().sort_values(by = 'Global_Sales', ascending = False).reset_index().head(5)
selling_platform
```

```python
# Visualization
fig, ax = plt.subplots(figsize=(25, 8))
plt.title("Top 5 best selling platforms",fontsize=25)
sns.barplot(data=selling_platform, x = "Platform", y = "Global_Sales")
```

### Observation

  * Playstation 2 are the leading selling platforms with over 1233.46 (in millions) followed by X360, PS3, Wii and DS.

### Top 10 best selling games of all time

```python
b_s_game = df[['Name', 'Global_Sales','Platform']].groupby(['Name']
                                                          ).sum().sort_values(by = 
                                                        'Global_Sales', ascending = False).reset_index().head(10)
b_s_game
```

```python
# Visualization
fig, ax = plt.subplots(figsize=(30, 15))
plt.title("Top 10 best selling games of all time",fontsize=25)
sns.barplot(data=b_s_game, x = "Name", y = "Global_Sales")
```

### Total sales  trend between 1980 to 2020.

```python
sales_trend = df.groupby("Year")[["Global_Sales"]].sum().reset_index()
sales_trend
```

```python
# visualizing the total global sales trend between 1980 to 2020
fig, ax = plt.subplots(figsize=(20, 8))
plt.title("Global sales trend of video games over the year 1980 - 2016",fontsize=20)
sns.lineplot(sales_trend["Year"], sales_trend["Global_Sales"], marker = "o", color="black")
```

### Observation
   * From the above visualization, it could be seen that Video game sales starts increasing from `1995` to `2008` due to the advancement of technology in the gaming market because as at this year i.e `1995` upward, the video gaming market gained a lot of popularity and lot of people started developing interest in their products rather than when it first started in the `1980's`.
   * In `2008`, the video game sales reaches its peak and started declining since then and this could be as a result of the global financial crisis that happened around this year. Also, in `2008` gaming took a new turn through the invention of mobile video games which has reduced the sales of game consoles around the world because most of the video games can be easily downloaded on a smartphone which in turn reduced the purchase of game console.
